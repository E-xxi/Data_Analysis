## 일시 정리

```python
def clean_date_type(df):
    df["clean_date"] = pd.to_datetime(df["num_date_time"].str.split("_").str[1],format='%Y%m%d %H')

    # 년/월/일/시간 분리
    df["month"] = df["clean_date"].dt.month
    df["day"] = df["clean_date"].dt.day
    df["hour"] = df["clean_date"].dt.hour
    
    df['태양광용량(kW)'] = pd.to_numeric(df['태양광용량(kW)'], errors='coerce').fillna(0).astype(float)
    df['ESS저장용량(kWh)'] = pd.to_numeric(df['ESS저장용량(kWh)'], errors='coerce').fillna(0).astype(float)
    df['PCS용량(kW)'] = pd.to_numeric(df['PCS용량(kW)'], errors='coerce').fillna(0).astype(float)

    
    del df['clean_date']

clean_date_type(train)
clean_date_type(test)
```

## 파생변수 생성
- 기상 관련(체감온도, 불쾌지수, 냉방도일, 일조 강도/ 일사량 지수)  
- 건물 특성(냉방면적 비율, 태양광 발전 잠재량, 에너지저장장치활용)  
- 시간적 특성(요일/주말, 피크시간대)  
- 종합적 (냉방부하지수, 에너지 자급률)

```python
import numpy as np

def build_features_summer_with_type_cdh(
    df: pd.DataFrame,
    *,
    # 건물유형별 냉방 기준온도(°C) 매핑: 없는 유형은 default_base_c 사용
    cdh_base_map: dict = None,
    default_base_c: float = 26.0,
    peak_hours = (10, 11, 14, 15, 16, 17),
    # 건물번호별 별도 기준온도 우선 적용 (유형보다 우선)
    cdh_base_by_building: dict = None
) -> pd.DataFrame:

    out = df.copy()

    # --- 시간 컬럼 정리 ---
    out['일시'] = pd.to_datetime(out['일시'], errors='coerce')
    out['weekday'] = out['일시'].dt.weekday   # 월=0, 일=6
    out['is_weekend'] = out.get('weekday', 0).isin([5, 6]).astype('int8') if 'weekday' in out.columns else 0
    out['holiday'] = out['일시'].dt.normalize().isin(pd.to_datetime(["2024-06-06", "2024-08-15"])).astype('int8')

    # --- 기초 수치 ---
    T  = out['기온(°C)'].astype(float)
    RH = out['습도(%)'].astype(float).clip(0, 100)
    V  = out['풍속(m/s)'].astype(float)
    rain = out['강수량(mm)'].astype(float).clip(lower=0)

    # --- 건물유형/번호별 냉방 기준온도 결정 (벡터화) ---
    # 1) 유형 매핑으로 기본 설정
    if cdh_base_map is None:
        cdh_base_map = {}  # 사용자가 외부에서 넘겨주도록

    base_from_type = out['건물유형'].map(cdh_base_map).astype(float)
    base = base_from_type.fillna(default_base_c)

    # 2) 건물번호별 오버라이드가 있으면 그것을 최우선 적용
    if cdh_base_by_building:
        # map 후 notna인 곳만 덮어쓰기
        base_override = out['건물번호'].map(cdh_base_by_building).astype(float)
        base = base_override.combine_first(base)

    out['cdh_base_used'] = base  # 실제 사용된 기준온도 기록(디버깅/설명용)

    # --- 여름 핵심 지표 ---
    # 불쾌지수(Thom DI)
    out['DI'] = 0.81*T + 0.01*RH*(0.99*T - 14.3) + 46.3

    # 온습도 지수(THI)
    out['THI'] = (9/5*T) - 0.55*(1 - RH/100.0)*((9/5*T) - 26) + 32

    # 냉방도일(CDH, 시점별) — 유형/건물별 기준온도 적용
    out['CDH'] = (T - out['cdh_base_used']).clip(lower=0)

    # --- 건물 특성/용량 ---
    area = out['연면적(m2)'].replace(0, np.nan)
    cool_area = out['냉방면적(m2)'].replace(0, np.nan)
    pv_kw = out['태양광용량(kW)'].astype(float)
    ess_kwh = out['ESS저장용량(kWh)'].astype(float)
    pcs_kw = out['PCS용량(kW)'].astype(float)

    out['cool_area_ratio'] = (cool_area / area).clip(0, 1)
    out['kWh_per_m2'] = np.nan  # test에는 타깃이 없으니 자리만(학습시 train에서 채워짐)
    out['ess_hours'] = np.where(pcs_kw > 0, ess_kwh / pcs_kw, np.nan)

    # --- 시간/계절성 & 상호작용 ---
    out['is_peak_hour'] = out['hour'].isin(peak_hours).astype('int8')
    out['hour_sin']  = np.sin(2*np.pi*out['hour']/24)
    out['hour_cos']  = np.cos(2*np.pi*out['hour']/24)
    out['month_sin'] = np.sin(2*np.pi*out['month']/12)
    out['month_cos'] = np.cos(2*np.pi*out['month']/12)

    out['is_rain']  = (rain > 0).astype('int8')
    out['T_x_RH']   = T * RH
    out['T2']       = T**2
    out['RH2']      = RH**2
    out['T_x_rain'] = T * out['is_rain']
    out['wind_level'] = pd.cut(V, bins=[-np.inf,1,3,6,10,np.inf], labels=[0,1,2,3,4]).astype(int)

    # --- 종합 냉방부하지수(모델용 지표) ---
    out['cooling_load_index'] = out['DI'] * out['cool_area_ratio'].fillna(0) + out['CDH']

    # --- 건물유형 원-핫 ---
    if '건물유형' in out.columns:
        out = pd.concat([out, pd.get_dummies(out['건물유형'], prefix='btype', dummy_na=False)], axis=1)

    # --- 수치 안정화 ---
    numeric_cols = out.select_dtypes(include=[np.number]).columns
    out[numeric_cols] = out[numeric_cols].replace([np.inf, -np.inf], np.nan)

    return out
```

```python
# 건물별 평균 실내 기온(예: 센서값 있거나 추정 가능)
building_avg_temp = train.groupby("건물번호")["기온(°C)"].mean()

# 기준온도 = 평균값 - 2
cdh_base_by_building = (building_avg_temp - 2).to_dict()

train_fe = build_features_summer_with_type_cdh(
    train,
    default_base_c=26.0,
    cdh_base_by_building=cdh_base_by_building
)
test_fe = build_features_summer_with_type_cdh(
    test,
    default_base_c=26.0,
    cdh_base_by_building=cdh_base_by_building
)
```
