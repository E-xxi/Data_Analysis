```python
test = pd.read_csv('Data/test_c20729.csv')
test
```

```python
train.drop(['num_date_time','일자'],axis=1, inplace = True)
```

```python
train = train[['건물유형', '건물번호', '기온(°C)', '강수량(mm)', '풍속(m/s)', '습도(%)', '일조(hr)', '일사(MJ/m2)',
       '전력소비량(kWh)', '연면적(m2)', '냉방면적(m2)', '태양광용량(kW)', 'ESS저장용량(kWh)',
       'PCS용량(kW)', '일시', '월', '일', '요일', '시간', '월_sin', '월_cos', 
               '요일_sin', '요일_cos', '시간_sin', '시간_cos']]
```

```python
import numpy as np

# 일의 사인/코사인
train['일_sin'] = np.sin(2 * np.pi * train['일'] / 31)
train['일_cos'] = np.cos(2 * np.pi * train['일'] / 31)

test['일_sin'] = np.sin(2 * np.pi * test['일'] / 31)
test['일_cos'] = np.cos(2 * np.pi * test['일'] / 31)
```

### 건물 종류에 따라서 타겟인코딩 시키자 min/max 값으로 

```python
train['건물유형_max'] = train.groupby('건물유형')['전력소비량(kWh)'].transform('max')
train['건물유형_mean'] = train.groupby('건물유형')['전력소비량(kWh)'].transform('mean')
```

```python
# test만 merge로 생성 (train에는 merge 금지)
test = test.merge(
    train.groupby(['건물유형'])['전력소비량(kWh)']
         .agg(건물유형_max='max', 건물유형_mean='mean')
         .reset_index(),
    on=['건물유형'], how='left'
)
```

### 건물 종류와 번호에 따라서도 타겟 인코딩

```python
train['건물번호_max'] = train.groupby(['건물유형','건물번호'])['전력소비량(kWh)'].transform('max')
train['건물번호_mean'] = train.groupby(['건물유형','건물번호'])['전력소비량(kWh)'].transform('mean')
```

```python
# test만 merge로 생성 (train에는 merge 금지)
test = test.merge(
    train.groupby(['건물유형','건물번호'])['전력소비량(kWh)']
         .agg(건물번호_max='max', 건물번호_mean='mean')
         .reset_index(),
    on=['건물유형','건물번호'], how='left'
)
```

```python
#건물유형, 종류, 번호, 요일, 시간 min max
train['num/day/time_max'] = train.groupby(['건물유형','건물번호','요일','시간'])['전력소비량(kWh)'].transform('max')
train['num/day/time_mean'] = train.groupby(['건물유형','건물번호','요일','시간'])['전력소비량(kWh)'].transform('mean')
```

```python
# test만 merge로 생성 (train에는 merge 금지)
test = test.merge(
    train.groupby(['건물유형','건물번호','요일','시간'])['전력소비량(kWh)']
         .agg(num_day_time_max='max', num_day_time_mean='mean')
         .reset_index(),
    on=['건물유형','건물번호','요일','시간'], how='left'
)
```

```python
# 0) 안전 복사
test = test.copy()

# 1) 중복 키(시간/요일) 통일: 우선순위는 *_y -> *_x -> 원본
def coalesce_cols(df, targets, new_name):
    for c in targets:
        if c in df.columns:
            df[new_name] = df[new_name] if new_name in df.columns else df[c]
            df.loc[df[new_name].isna(), new_name] = df[c]
    # 원본 열들 제거
    for c in targets:
        if c in df.columns:
            df.drop(columns=c, inplace=True, errors='ignore')

# 시간/요일 합치기: 시간, 요일로 최종 통일
coalesce_cols(test, ['시간_y','시간_x','시간'], '시간')
coalesce_cols(test, ['요일_y','요일_x','요일'], '요일')

# 2) 건물유형 통계 중복 통일: *_y 우선, 없으면 *_x
def coalesce_pair(df, left, right, new):
    a = df[left] if left in df.columns else None
    b = df[right] if right in df.columns else None
    if a is not None and b is not None:
        df[new] = b.fillna(a)
    elif b is not None:
        df[new] = b
    elif a is not None:
        df[new] = a
    # 원본 제거
    for c in [left, right]:
        if c in df.columns:
            df.drop(columns=c, inplace=True, errors='ignore')

coalesce_pair(test, '건물유형_max_x',  '건물유형_max_y',  '건물유형_max')
coalesce_pair(test, '건물유형_mean_x', '건물유형_mean_y', '건물유형_mean')

# 3) 슬래시 있는 이름은 파이프라인 안정성을 위해 언더스코어로(필요 시)
if 'num/day/time_max' in test.columns:
    test.rename(columns={'num/day/time_max':'num_day_time_max'}, inplace=True)
if 'num/day/time_mean' in test.columns:
    test.rename(columns={'num/day/time_mean':'num_day_time_mean'}, inplace=True)

# 4) holiday 타입 train과 맞추기 (train이 bool이면 bool로)
if 'holiday' in test.columns and 'holiday' in train.columns:
    if pd.api.types.is_bool_dtype(train['holiday']) and not pd.api.types.is_bool_dtype(test['holiday']):
        # 0/1, int, float → bool 변환
        test['holiday'] = test['holiday'].astype('float').fillna(0).astype('int').astype('bool')

# 5) 불필요/중복 유사 컬럼 정리(있으면 제거)
drop_candidates = [
    '시간_x','시간_y','요일_x','요일_y',
    '건물유형_max_x','건물유형_max_y',
    '건물유형_mean_x','건물유형_mean_y'
]
test.drop(columns=[c for c in drop_candidates if c in test.columns], inplace=True, errors='ignore')

# 6) 최종 컬럼 집합 맞추기 (타깃 제외)
TARGET = '전력소비량(kWh)'
train_features = train.drop(columns=[TARGET])
# test에 없는 열은 생성해서 채우고, 있는 열은 그대로 사용
for col in train_features.columns:
    if col not in test.columns:
        # 기본 결측 채움 규칙: 수치=0, 범주형='unknown', bool=False
        if pd.api.types.is_numeric_dtype(train_features[col]):
            test[col] = 0
        elif pd.api.types.is_bool_dtype(train_features[col]):
            test[col] = False
        else:
            test[col] = 'unknown'

# train에는 없는데 test에만 있는 열은 버림
extra_cols = [c for c in test.columns if c not in train_features.columns]
test.drop(columns=extra_cols, inplace=True, errors='ignore')

# 7) 순서까지 train과 동일하게
test = test[train_features.columns]

# --- 확인용 출력 ---
print("정렬 후 차이(있으면 안 됨):")
print("train-only :", set(train_features.columns) - set(test.columns))
print("test-only  :", set(test.columns) - set(train_features.columns))
print(test.dtypes.head(20))
```
